### =================Principle - Component-Analysis ==========================================

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
wine=pd.read_csv("wine.csv")
```

```python
wine
```

```python
wine2=wine.iloc[:,1:]
```

```python
wine2
```

```python
wine2.describe()
```

```python
wine3=wine2.values
```

```python
wine3
```

```python
wine4=scale(wine3)
```

```python
wine4
```

```python
pca=PCA(n_components=13)
```

```python
wine_pca=pca.fit_transform(wine4)
```

```python
wine_pca
```

```python
pca.components_
```

```python
variance=pca.explained_variance_ratio_
```

```python
variance
```

```python
variance1=np.cumsum(np.round(variance,4)*100)
```

```python
variance1
```

```python
get_ipython().run_line_magic('matplotlib','inline')
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(variance1)
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(25,8))
Dendogram=sch.dendrogram(sch.linkage(wine4,'complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(wine4),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
wine5=wine.copy()
wine5['clustersid']=hclusters.labels_
wine5
```

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
abc=[]
for i in range(1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(wine4)
    abc.append(kmeans.inertia_)
```

```python
plt.plot(range(1,6),abc)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel("WCSS")
plt.show()
```

```python
clusters3=KMeans(3,random_state=30).fit(wine4)
clusters3
```

```python
clusters3.labels_
```

```python
wine6=wine.copy()
wine6['cluster3sid']=clusters3.labels_
```

```python
wine6
```
